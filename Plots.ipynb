```python
election_cost = [[],[],[]]
election_count = [[],[],[]]
broadcast_cost = [[],[],[]]
broadcast_count = [[],[],[]]
query_cost = [[],[],[]]
query_count = [[],[],[]]

for i in range(3,25):
    baselinefn = 'data/baseline'+str(i)+'.out'
    mstfn = 'data/mst'+str(i)+'.out'
    shortestpathfn = 'data/shortestpath'+str(i)+'.out'
    for j in range(0, 3):
        if j == 0:
            fn = baselinefn
        elif j == 1:
            fn = mstfn
        else:
            fn = shortestpathfn
        with open(fn,'r') as f:
            ln = 1
            for line in f:
                temp = line.split(": ")
                value = float(temp[1])
                if (ln == 1):
                    election_cost[j].append(value)
                elif (ln == 2):
                    election_count[j].append(value)
                elif (ln == 3):
                    broadcast_cost[j].append(value)
                elif (ln == 4):
                    broadcast_count[j].append(value)
                elif (ln == 5):
                    query_cost[j].append(value)
                else:
                    query_count[j].append(value)
                ln = ln + 1
```
